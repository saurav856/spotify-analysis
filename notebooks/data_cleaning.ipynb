```python
import pandas as pd

df = pd.read_csv('../data/raw/dataset.csv')
print(df.shape)
print(df.columns.tolist())
print(df.head(2))
```

```python
print(df.isnull().sum())
print(f"\nDuplicate rows: {df.duplicated().sum()}")
print(f"Duplicate track_ids: {df['track_id'].duplicated().sum()}")
```

```python
# drop nulls
df = df.dropna()

# drop unnamed column
df = df.drop(columns=['Unnamed: 0'])

# convert duration from ms to seconds
df['duration_sec'] = df['duration_ms'] / 1000

# extract decade from... wait this dataset has no year column
print(df.columns.tolist())
```

```python
print(df.dtypes)
print(df.head(1).to_dict())
```

```python
# drop nulls
df = df.dropna()

# drop duration_ms since we have duration_sec now
df = df.drop(columns=['duration_ms'])

print(df.shape)
print(df.dtypes)
```

```python
df = df.applymap(lambda x: x.encode('utf-8', 'ignore').decode('utf-8') if isinstance(x, str) else x)
df.to_csv('../data/cleaned/spotify_cleaned.csv', index=False, encoding='utf-8')
print("Saved successfully")
```
